```python
def mostrar_imagenes_grid(lista, cols=4, figsize=None, suptitle=None):
    pares = []
    for item in lista:
        if isinstance(item, dict):
            if 'imagen' in item and 'titulo' in item:
                pares.append((item['imagen'], item['titulo']))
            else:
                for k, v in item.items():
                    if isinstance(k, str) and isinstance(v, np.ndarray):
                        pares.append((v, k))  # {'Titulo': imagen}
                    else:
                        # Fallback genérico
                        pares.append((v, str(k)))
        elif isinstance(item, (tuple, list)) and len(item) >= 2:
            pares.append((item[0], item[1]))
        else:
            pares.append((item, ''))

    n = len(pares)
    if n == 0:
        raise ValueError('Lista vacía')

    rows = math.ceil(n / cols)
    if figsize is None:
        figsize = (4 * cols, 3.5 * rows)

    fig, axes = plt.subplots(rows, cols, figsize=figsize)
    axes = np.array(axes).reshape(rows, cols)

    for i, (imagen, titulo) in enumerate(pares):
        ax = axes.flat[i]
        if imagen is None or getattr(imagen, 'size', 0) == 0:
            ax.set_title(f'{titulo} (vacía)')
            ax.axis('off')
            continue
        if getattr(imagen, 'ndim', 0) == 2:
            img, cmap = imagen, 'gray'
        else:
            img, cmap = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB), None
        ax.imshow(img, cmap=cmap)
        if titulo:
            ax.set_title(titulo)
        ax.axis('off')

    for j in range(n, rows * cols):
        axes.flat[j].axis('off')

    if suptitle:
        fig.suptitle(suptitle)
    plt.tight_layout()
    if suptitle:
        plt.subplots_adjust(top=0.92)
    return fig, axes
```

```python
def graficar_blur_fm_frames(xs, fm_values, min_idx, max_idx, min_val, max_val):
    plt.figure(figsize=(10, 4))
    plt.plot(xs, fm_values, label="frequency_blur_fm")
    plt.scatter([min_idx], [min_val], color='red', label=f"min={min_val:.4f} (f{min_idx})")
    plt.scatter([max_idx], [max_val], color='green', label=f"max={max_val:.4f} (f{max_idx})")
    plt.xlabel("Frame")
    plt.ylabel("FM")
    plt.title("frequency_blur_fm por frame")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def central_crop(frame, ROI_from_center):
    """
    Toma el <ROI_from_center> % central de cada dimensión (ancho y alto) por defecto.
    ROI_from_center: ROI_from_centerción por lado en (0, 1], p.ej., 0.1 => 10% del ancho y 10% del alto.
    """
    assert 0 < ROI_from_center <= 1.0
    h, w = frame.shape[:2]
    ch = max(1, int(round(h * ROI_from_center)))
    cw = max(1, int(round(w * ROI_from_center)))
    y1 = (h - ch) // 2
    x1 = (w - cw) // 2
    return frame[y1:y1 + ch, x1:x1 + cw]
```

```python
def get_cap_blur_fm_information(cap, metrica_enfoque, sharp, ROI_from_center):
    n_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fm_values = []
    min_idx = None
    max_idx = None
    min_val = None
    max_val = None
    min_frame = None
    max_frame = None
    i = 0
    while True:
        ok, frame = cap.read()
        if not ok or frame is None:
            break
        roi = central_crop(frame, ROI_from_center)
        if sharp:
            roi = unsharp_mask(roi, k=1, sigma=1)
        v = metrica_enfoque(roi)
        fm_values.append(v)
        if min_val is None or v < min_val:
            min_val = v
            min_idx = i
            min_frame = roi.copy()
        if max_val is None or v > max_val:
            max_val = v
            max_idx = i
            max_frame = roi.copy()
        i += 1
    return fm_values, min_idx, max_idx, min_val, max_val, min_frame, max_frame
```

```python
def graficar_fm_video(video_path, metrica_enfoque=frequency_blur_fm, sharp=False, ROI_from_center=1):
    cap = cv2.VideoCapture(str(video_path), cv2.CAP_FFMPEG)
    if not cap.isOpened():
        cap = cv2.VideoCapture(str(video_path))
        if not cap.isOpened():
            raise RuntimeError(f"No se pudo abrir el video: {video_path}")
    fm_values, min_idx, max_idx, min_val, max_val, min_frame, max_frame = get_cap_blur_fm_information(cap, metrica_enfoque, sharp, ROI_from_center)
    cap.release()
    if len(fm_values) == 0:
        raise RuntimeError("No se pudieron leer frames del video.")
    xs = list(range(len(fm_values)))
    graficar_blur_fm_frames(xs, fm_values, min_idx, max_idx, min_val, max_val)

    lista = [
        (max_frame, f"Máximo FM (frame {max_idx}, {max_val:.4f})"),
        (min_frame, f"Mínimo FM (frame {min_idx}, {min_val:.4f})"),
    ]
    mostrar_imagenes_grid(lista, cols=2, suptitle="Frames de máximo y mínimo FM")
    return {
        "fm_values": fm_values,
        "min_idx": min_idx,
        "min_val": float(min_val),
        "max_idx": max_idx,
        "max_val": float(max_val),
    }
```

### Funciones Auxiliares Resolucion 1

```python
def frequency_blur_fm(image: np.ndarray) -> float:
    if image.ndim == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image
    gray = np.asarray(gray, dtype=np.float32)
    if gray.ndim != 2 or gray.size == 0:
        return 0.0
    rows, cols = gray.shape
    if rows == 0 or cols == 0:
        return 0.0
    # Step 1: Compute F which is the Fourier Transform representation of image I 
    F = np.fft.fft2(gray)
    # Step 2: Find Fc which is obtained by shifting the origin of F to centre. 
    Fc = np.fft.fftshift(F)
    # Step 3: Calculate AF = abs (Fc) where AF is the absolute value of the centered Fourier transform of image I. 
    AF = np.abs(Fc)
    # Step 4: Calculate M = max (AF) where M is the maximum value of the frequency component in F
    M = float(AF.max()) if AF.size else 0.0
    if M <= 0.0:
        return 0.0
    # Step 5: Calculate TH = the total number of pixels in F whose pixel value > thres, where thres = M/1000
    thres = M / 1000.0
    TH = int(np.count_nonzero(AF > thres))
    # Step 6: Calculate Image Quality measure (FM) from equation FM = TH / (rows*cols)
    FM = TH / float(rows * cols)
    return float(FM)
```

### Funciones Auxiliares Resolucion 2

```python
def lap1_energy(image: np.ndarray) -> float:
    if image.ndim == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image
    gray = np.asarray(gray, dtype=np.float32)
    if gray.ndim != 2 or gray.size == 0:
        return 0.0
    lap = cv2.Laplacian(gray, cv2.CV_32F, ksize=3)
    energy = float(np.sum(lap * lap))
    return energy
```

### Funciones Auxiliares Unsharp Mask

```python
def unsharp_mask(image: np.ndarray, k: float = 1.5, sigma: float = 1.0) -> np.ndarray:
    if image is None or getattr(image, 'size', 0) == 0:
        return image
    orig_dtype = image.dtype
    img = image.astype(np.float32)
    blurred = cv2.GaussianBlur(img, (0, 0), sigmaX=sigma, sigmaY=sigma)
    sharpened = (1.0 + k) * img - k * blurred
    if np.issubdtype(orig_dtype, np.integer):
        sharpened = np.clip(sharpened, 0, 255).astype(orig_dtype)
    else:
        sharpened = sharpened.astype(np.float32)
    return sharpened
```

### Funciones Auxiliares Comparacion de Metricas

```python
def comparacion_metricas(res_1, res_2, metrica_enfoque_1, metrica_enfoque_2):
    # Comparación de métricas: frequency_blur_fm vs LAP1, normalizadas
    res_1_vals = res_1.get('fm_values')
    res_2_vals = res_2.get('fm_values')

    if res_1_vals is not None and res_2_vals is not None:
        res_1_vals = np.array(res_1_vals, dtype=float)
        res_2_vals = np.array(res_2_vals, dtype=float)

        # Normalización min-max para graficar en la misma escala
        def norm01(x):
            xmin, xmax = np.min(x), np.max(x)
            return (x - xmin) / (xmax - xmin + 1e-12)

        res_1_n = norm01(res_1_vals)
        res_2_n = norm01(res_2_vals)

        xs = np.arange(len(res_1_n))
        plt.figure(figsize=(10,4))
        plt.plot(xs, res_1_n, label=f'{metrica_enfoque_1} (norm)')
        plt.plot(xs, res_2_n, label=f'{metrica_enfoque_2} (norm)')
        plt.xlabel('Frame')
        plt.ylabel('Valor normalizado [0,1]')
        plt.title('Comparación de métricas de enfoque (normalizadas)')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
    else:
        print('No se encontraron valores para comparar')
```
